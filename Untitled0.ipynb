```python
import os
import json
import librosa
import soundfile as sf
import pandas as pd
from pathlib import Path

os.makedirs("data/processed_audio", exist_ok=True)

# ─── PREPROCESSING ───────────────────────────────────────────
# For each recording:
#   1. Load full audio
#   2. Convert 44100Hz → 16000Hz
#   3. Split into segments using start/end times from JSON
#   4. Filter bad segments
#   5. Save each segment as a small .wav file
#   6. Build a master CSV with (audio_path, text)

TARGET_SR = 16000
MIN_DURATION = 1.0   # skip segments shorter than 1 second
MAX_DURATION = 30.0  # skip segments longer than 30 seconds (Whisper limit)

records = []
skipped = 0
total = 0

audio_files = list(Path("data/audio").glob("*.wav"))
print(f"Processing {len(audio_files)} audio files...\n")

for audio_path in audio_files:
    rec_id = audio_path.stem  # e.g. "825780"
    trans_path = Path(f"data/transcriptions/{rec_id}.json")

    # Skip if no transcription
    if not trans_path.exists():
        print(f"  ✗ No transcription for {rec_id}")
        continue

    # Load full audio at 16kHz directly (librosa resamples automatically)
    try:
        y, sr = librosa.load(str(audio_path), sr=TARGET_SR, mono=True)
    except Exception as e:
        print(f"  ✗ Failed to load {rec_id}: {e}")
        continue

    # Load transcription segments
    with open(trans_path, encoding='utf-8') as f:
        segments = json.load(f)

    for seg in segments:
        total += 1
        start = seg['start']
        end   = seg['end']
        text  = seg['text'].strip()
        duration = end - start

        # ── Filters ──────────────────────────────────────────
        if not text:
            skipped += 1
            continue
        if duration < MIN_DURATION or duration > MAX_DURATION:
            skipped += 1
            continue

        # ── Slice audio segment ───────────────────────────────
        start_sample = int(start * TARGET_SR)
        end_sample   = int(end   * TARGET_SR)
        segment_audio = y[start_sample:end_sample]

        # ── Save segment audio ────────────────────────────────
        seg_filename = f"{rec_id}_{int(start*100):08d}.wav"
        seg_path = f"data/processed_audio/{seg_filename}"
        sf.write(seg_path, segment_audio, TARGET_SR)

        # ── Store record ──────────────────────────────────────
        records.append({
            "audio_path"  : seg_path,
            "text"        : text,
            "duration"    : round(duration, 2),
            "recording_id": rec_id,
            "speaker_id"  : seg.get('speaker_id', ''),
            "start"       : start,
            "end"         : end
        })

# ─── Save master CSV ─────────────────────────────────────────
df = pd.DataFrame(records)
df.to_csv("data/dataset.csv", index=False, encoding='utf-8')

print(f"=== Preprocessing Complete ===")
print(f"Total segments found  : {total}")
print(f"Segments skipped      : {skipped}")
print(f"Segments kept         : {len(records)}")
print(f"Total audio duration  : {df['duration'].sum()/3600:.2f} hours")
print(f"\nSample rows:")
print(df[['audio_path','text','duration']].head(3).to_string())
```

```python
import pandas as pd
import torch
from datasets import Dataset, Audio
from transformers import WhisperProcessor

# ─── Load our CSV ────────────────────────────────────────────
df = pd.read_csv("data/dataset.csv")
print(f"Total samples: {len(df)}")

# Keep only what we need
df = df[['audio_path', 'text']].rename(columns={'audio_path': 'audio'})

# ─── Train / Test split (90% train, 10% test) ────────────────
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.1, random_state=42)
print(f"Train: {len(train_df)} | Test: {len(test_df)}")

# ─── Convert to HuggingFace Dataset ──────────────────────────
train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True))
test_dataset  = Dataset.from_pandas(test_df.reset_index(drop=True))

# Tell HuggingFace these are audio files at 16kHz
train_dataset = train_dataset.cast_column("audio", Audio(sampling_rate=16000))
test_dataset  = test_dataset.cast_column("audio",  Audio(sampling_rate=16000))

print(f"\nSample entry:")
print(train_dataset[0])
```

```python
from transformers import WhisperProcessor
import numpy as np

# ─── Load Whisper processor ──────────────────────────────────
# This handles both audio features + text tokenization
processor = WhisperProcessor.from_pretrained(
    "openai/whisper-small",
    language="hindi",
    task="transcribe"
)

print("Processor loaded!")

# ─── Feature preparation function ───────────────────────────
def prepare_dataset(batch):
    # Load audio
    audio = batch["audio"]

    # Extract audio array and sampling rate
    audio_array = audio["array"]
    sampling_rate = audio["sampling_rate"]

    # Convert audio to log-mel spectrogram (what Whisper understands)
    batch["input_features"] = processor.feature_extractor(
        audio_array,
        sampling_rate=sampling_rate
    ).input_features[0]

    # Tokenize the Hindi text into label IDs
    batch["labels"] = processor.tokenizer(batch["text"]).input_ids

    return batch

# ─── Apply to both splits ────────────────────────────────────
print("Processing train set...")
train_dataset_processed = train_dataset.map(
    prepare_dataset,
    remove_columns=["audio", "text"],
    num_proc=1
)

print("Processing test set...")
test_dataset_processed = test_dataset.map(
    prepare_dataset,
    remove_columns=["audio", "text"],
    num_proc=1
)

print(f"\n=== Done ===")
print(f"Train features: {train_dataset_processed[0]['input_features'].shape}")
print(f"Sample label IDs: {train_dataset_processed[0]['labels'][:10]}")
print(f"Decoded back: {processor.tokenizer.decode(train_dataset_processed[0]['labels'])}")
```

```python
import torch
from dataclasses import dataclass
from typing import Any, Dict, List, Union
from transformers import WhisperForConditionalGeneration, Seq2SeqTrainingArguments, Seq2SeqTrainer
import evaluate

# ─── 1. Data Collator ────────────────────────────────────────
# This pads each batch to same length during training
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        # Pad input features
        input_features = [{"input_features": f["input_features"]} for f in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

        # Pad labels
        label_features = [{"input_ids": f["labels"]} for f in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")

        # Replace padding token id with -100 (ignored in loss calculation)
        labels = labels_batch["input_ids"].masked_fill(
            labels_batch.attention_mask.ne(1), -100
        )

        # Remove start token if present
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]

        batch["labels"] = labels
        return batch

data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)
print("✓ Data collator ready!")

# ─── 2. Load Whisper-small model ─────────────────────────────
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
model.generation_config.language = "hindi"
model.generation_config.task = "transcribe"
model.generation_config.forced_decoder_ids = None
print("✓ Model loaded!")

# ─── 3. WER metric ───────────────────────────────────────────
wer_metric = evaluate.load("wer")

def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids

    # Replace -100 with pad token
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id

    # Decode predictions and labels
    pred_str  = processor.tokenizer.batch_decode(pred_ids,  skip_special_tokens=True)
    label_str = processor.tokenizer.batch_decode(label_ids, skip_special_tokens=True)

    wer = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"wer": round(wer, 4)}

print("✓ Metrics ready!")

# ─── 4. Training Arguments ───────────────────────────────────
training_args = Seq2SeqTrainingArguments(
    output_dir="./whisper-small-hindi",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2,
    learning_rate=1e-5,
    warmup_steps=100,
    max_steps=500,
    gradient_checkpointing=True,
    fp16=True,
    eval_strategy="steps",
    eval_steps=100,
    save_steps=100,
    logging_steps=25,
    predict_with_generate=True,
    generation_max_length=225,
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False,
    report_to="none"
)

print("✓ Training arguments set!")
print(f"\nTraining for {training_args.max_steps} steps")
print(f"Batch size: {training_args.per_device_train_batch_size}")
print(f"Evaluating every {training_args.eval_steps} steps")
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset_processed,
    eval_dataset=test_dataset_processed,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    processing_class=processor.feature_extractor,
)

print("Starting training... this will take 20-30 minutes on Colab GPU")
print("You will see progress every 25 steps\n")

trainer.train()

print("\n✓ Training complete!")
```

```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
from dataclasses import dataclass
from typing import Any, Dict, List, Union
import torch, evaluate
import numpy as np

# ── Load processor + model ────────────────────────────────────
processor = WhisperProcessor.from_pretrained(
    "openai/whisper-small", language="hindi", task="transcribe"
)
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
model.generation_config.language = "hindi"
model.generation_config.task = "transcribe"
model.generation_config.forced_decoder_ids = None
print("✓ Processor and model loaded!")

# ── Build HuggingFace dataset ─────────────────────────────────
df_proc = pd.read_csv("data/dataset.csv")
train_df, test_df = train_test_split(df_proc[['audio','text']], test_size=0.1, random_state=42)

train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True)).cast_column("audio", Audio(sampling_rate=16000))
test_dataset  = Dataset.from_pandas(test_df.reset_index(drop=True)).cast_column("audio",  Audio(sampling_rate=16000))

def prepare_dataset(batch):
    audio = batch["audio"]
    batch["input_features"] = processor.feature_extractor(
        audio["array"], sampling_rate=audio["sampling_rate"]
    ).input_features[0]
    batch["labels"] = processor.tokenizer(batch["text"]).input_ids
    return batch

train_dataset_processed = train_dataset.map(prepare_dataset, remove_columns=["audio","text"], num_proc=1)
test_dataset_processed  = test_dataset.map(prepare_dataset,  remove_columns=["audio","text"], num_proc=1)
print(f"✓ Dataset ready! Train: {len(train_dataset_processed)} | Test: {len(test_dataset_processed)}")

# ── Data collator ─────────────────────────────────────────────
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any
    def __call__(self, features):
        input_features = [{"input_features": f["input_features"]} for f in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")
        label_features = [{"input_ids": f["labels"]} for f in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")
        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]
        batch["labels"] = labels
        return batch

data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)

# ── Metrics ───────────────────────────────────────────────────
wer_metric = evaluate.load("wer")
def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id
    pred_str  = processor.tokenizer.batch_decode(pred_ids,  skip_special_tokens=True)
    label_str = processor.tokenizer.batch_decode(label_ids, skip_special_tokens=True)
    return {"wer": round(wer_metric.compute(predictions=pred_str, references=label_str), 4)}

print("✓ Collator and metrics ready!")
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./whisper-small-hindi",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2,
    learning_rate=1e-5,
    warmup_steps=100,
    max_steps=500,
    gradient_checkpointing=True,
    fp16=True,
    eval_strategy="steps",
    eval_steps=100,
    save_steps=100,
    logging_steps=25,
    predict_with_generate=True,
    generation_max_length=225,
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False,
    report_to="none"
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset_processed,
    eval_dataset=test_dataset_processed,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    processing_class=processor.feature_extractor,
)

print("🚀 Starting training on GPU T4...")
trainer.train()
print("\n✓ Training complete!")
```

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor
import torch
import evaluate
import pandas as pd
from datasets import Dataset, Audio
from sklearn.model_selection import train_test_split

wer_metric = evaluate.load("wer")

# ── Use our own test split for evaluation ─────────────────────
# We already have 493 test samples - this is valid evaluation!
df_proc = pd.read_csv("data/dataset.csv")
train_df, test_df = train_test_split(
    df_proc[['audio','text']], test_size=0.1, random_state=42
)
test_dataset = Dataset.from_pandas(
    test_df.reset_index(drop=True)
).cast_column("audio", Audio(sampling_rate=16000))

print(f"Test samples: {len(test_dataset)}")

# ── Load BASELINE Whisper-small ───────────────────────────────
print("\nLoading baseline model...")
baseline_processor = WhisperProcessor.from_pretrained(
    "openai/whisper-small", language="hindi", task="transcribe"
)
baseline_model = WhisperForConditionalGeneration.from_pretrained(
    "openai/whisper-small"
).to("cuda")
baseline_model.eval()
print("✓ Baseline model loaded!")

# ── Run baseline evaluation ───────────────────────────────────
print("\nEvaluating baseline... (this takes ~5 mins)")

baseline_preds = []
baseline_refs  = []

for i, sample in enumerate(test_dataset):
    audio_array = sample["audio"]["array"]
    reference   = sample["text"]

    # Prepare input
    inputs = baseline_processor.feature_extractor(
        audio_array, sampling_rate=16000, return_tensors="pt"
    ).input_features.to("cuda")

    # Generate transcription
    with torch.no_grad():
        predicted_ids = baseline_model.generate(inputs)

    prediction = baseline_processor.tokenizer.batch_decode(
        predicted_ids, skip_special_tokens=True
    )[0]

    baseline_preds.append(prediction)
    baseline_refs.append(reference)

    if (i+1) % 50 == 0:
        print(f"  [{i+1}/{len(test_dataset)}] done...")

baseline_wer = wer_metric.compute(
    predictions=baseline_preds,
    references=baseline_refs
)
print(f"\n✓ Baseline WER: {baseline_wer:.4f} ({baseline_wer*100:.2f}%)")
```

```python
# ── Load FINE-TUNED model from Google Drive ───────────────────
print("Loading fine-tuned model...")
finetuned_processor = WhisperProcessor.from_pretrained(
    "/content/drive/MyDrive/hindi_asr/whisper-small-hindi-finetuned"
)
finetuned_model = WhisperForConditionalGeneration.from_pretrained(
    "/content/drive/MyDrive/hindi_asr/whisper-small-hindi-finetuned"
).to("cuda")
finetuned_model.generation_config.language = "hindi"
finetuned_model.generation_config.task = "transcribe"
finetuned_model.eval()
print("✓ Fine-tuned model loaded!")

# ── Run fine-tuned evaluation ─────────────────────────────────
print("\nEvaluating fine-tuned model... (this takes ~5 mins)")

finetuned_preds = []
finetuned_refs  = []

for i, sample in enumerate(test_dataset):
    audio_array = sample["audio"]["array"]
    reference   = sample["text"]

    inputs = finetuned_processor.feature_extractor(
        audio_array, sampling_rate=16000, return_tensors="pt"
    ).input_features.to("cuda")

    with torch.no_grad():
        predicted_ids = finetuned_model.generate(
            inputs,
            language="hi",
            task="transcribe"
        )

    prediction = finetuned_processor.tokenizer.batch_decode(
        predicted_ids, skip_special_tokens=True
    )[0]

    finetuned_preds.append(prediction)
    finetuned_refs.append(reference)

    if (i+1) % 50 == 0:
        print(f"  [{i+1}/{len(test_dataset)}] done...")

finetuned_wer = wer_metric.compute(
    predictions=finetuned_preds,
    references=finetuned_refs
)
print(f"\n✓ Fine-tuned WER: {finetuned_wer:.4f} ({finetuned_wer*100:.2f}%)")

# ── Also fix baseline with forced Hindi ──────────────────────
print("\nRe-evaluating baseline WITH Hindi forced (fair comparison)...")
baseline_preds_hindi = []

for i, sample in enumerate(test_dataset):
    audio_array = sample["audio"]["array"]
    inputs = baseline_processor.feature_extractor(
        audio_array, sampling_rate=16000, return_tensors="pt"
    ).input_features.to("cuda")

    with torch.no_grad():
        predicted_ids = baseline_model.generate(
            inputs,
            language="hi",
            task="transcribe"
        )

    prediction = baseline_processor.tokenizer.batch_decode(
        predicted_ids, skip_special_tokens=True
    )[0]
    baseline_preds_hindi.append(prediction)

    if (i+1) % 50 == 0:
        print(f"  [{i+1}/{len(test_dataset)}] done...")

baseline_wer_hindi = wer_metric.compute(
    predictions=baseline_preds_hindi,
    references=finetuned_refs
)
print(f"\n✓ Baseline WER (Hindi forced): {baseline_wer_hindi:.4f} ({baseline_wer_hindi*100:.2f}%)")

# ── Final comparison table ────────────────────────────────────
print("\n" + "="*55)
print("           WER COMPARISON TABLE")
print("="*55)
print(f"{'Model':<35} {'WER':>10}")
print("-"*55)
print(f"{'Baseline Whisper-small (no lang)':<35} {baseline_wer*100:>9.2f}%")
print(f"{'Baseline Whisper-small (Hindi forced)':<35} {baseline_wer_hindi*100:>9.2f}%")
print(f"{'Fine-tuned Whisper-small (ours)':<35} {finetuned_wer*100:>9.2f}%")
print("="*55)
improvement = (baseline_wer_hindi - finetuned_wer) / baseline_wer_hindi * 100
print(f"\nImprovement over baseline: {improvement:.1f}%")
```

```python
import pandas as pd

# ── Build error dataframe ─────────────────────────────────────
results_df = pd.DataFrame({
    "reference"  : finetuned_refs,
    "prediction" : finetuned_preds,
})

# Calculate per-utterance WER
def utterance_wer(ref, pred):
    ref_words  = ref.strip().split()
    pred_words = pred.strip().split()
    if len(ref_words) == 0:
        return 0.0
    # Simple word error count
    ref_len  = len(ref_words)
    pred_len = len(pred_words)
    # Use evaluate for accuracy
    try:
        score = wer_metric.compute(predictions=[pred], references=[ref])
    except:
        score = 1.0
    return round(score, 4)

print("Computing per-utterance WER... (takes 1-2 mins)")
results_df['wer'] = results_df.apply(
    lambda row: utterance_wer(row['reference'], row['prediction']), axis=1
)

# ── Filter only errors (WER > 0) ─────────────────────────────
errors_df = results_df[results_df['wer'] > 0].copy()
errors_df = errors_df.sort_values('wer', ascending=False).reset_index(drop=True)

print(f"\nTotal utterances : {len(results_df)}")
print(f"Utterances with errors: {len(errors_df)}")
print(f"Perfect predictions   : {len(results_df) - len(errors_df)}")

# ── Systematic sampling strategy ─────────────────────────────
# Strategy: Stratified by severity
# Low error    : WER 0.01 - 0.30  (minor mistakes)
# Medium error : WER 0.31 - 0.70  (moderate mistakes)
# High error   : WER 0.71 - 1.00  (severe mistakes)
# Very high    : WER > 1.00       (hallucinations/complete failures)

low    = errors_df[errors_df['wer'].between(0.01, 0.30)]
medium = errors_df[errors_df['wer'].between(0.31, 0.70)]
high   = errors_df[errors_df['wer'].between(0.71, 1.00)]
very_high = errors_df[errors_df['wer'] > 1.00]

print(f"\n=== Error Severity Distribution ===")
print(f"Low    (WER 0-30%)   : {len(low)} utterances")
print(f"Medium (WER 31-70%)  : {len(medium)} utterances")
print(f"High   (WER 71-100%) : {len(high)} utterances")
print(f"Very High (WER >100%): {len(very_high)} utterances")

# Sample from each bucket (total ~25-30 samples)
import numpy as np
np.random.seed(42)

sampled = pd.concat([
    low.sample(min(8, len(low))),
    medium.sample(min(8, len(medium))),
    high.sample(min(5, len(high))),
    very_high.sample(min(4, len(very_high)))
]).reset_index(drop=True)

print(f"\n=== Sampled {len(sampled)} error utterances ===")
for i, row in sampled.iterrows():
    print(f"\n[{i+1}] WER: {row['wer']:.2f}")
    print(f"  REF : {row['reference']}")
    print(f"  PRED: {row['prediction']}")
```

```python
# ── Part e: Error Taxonomy ────────────────────────────────────
# Based on analysis of our 25 sampled errors

taxonomy = """
╔══════════════════════════════════════════════════════════════╗
║           ERROR TAXONOMY - FINE-TUNED WHISPER-SMALL         ║
╚══════════════════════════════════════════════════════════════╝

CATEGORY 1: PHONETIC CONFUSION (Most Frequent)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Similar sounding words get confused. Model mishears consonants/vowels.

Examples:
  [3] REF: ओलंपिक समय    → PRED: ओलंपिक सम्में
  [3] REF: गेम्स          → PRED: गेंज्ट
  [5] REF: सुन रहा        → PRED: सुम रहा
  [6] REF: संगीत          → PRED: संगीद
  [7] REF: कभी            → PRED: तभी
CAUSE: Hindi has many phonetically similar consonant pairs
       (स/श, त/ट, ब/व, ड/ड़). Model confuses these under noise.

CATEGORY 2: SPELLING VARIATION / SCHWA DELETION (Very Frequent)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Different valid spellings of same spoken word. Schwa (अ) dropped.

Examples:
  [2] REF: सपने           → PRED: सप्ने
  [2] REF: कार्य          → PRED: कारी
  [7] REF: दोबारा         → PRED: दुबारा
  [9] REF: किलियर         → PRED: किल्यर
  [12] REF: फिलहाल        → PRED: फिलाल
CAUSE: Conversational Hindi has informal spellings. Training data
       has inconsistent spellings confusing the model.

CATEGORY 3: SHORT UTTERANCE HALLUCINATION (Frequent)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
For very short utterances, model hallucinates extra words.

Examples:
  [21] REF: जी            → PRED: हां जी
  [23] REF: ओके           → PRED: ओकी है
  [24] REF: हाअ           → PRED: हां हां
  [25] REF: बट            → PRED: प्रोफमेंज उसका वैक्ट
  [18] REF: हां बिल्कुल   → PRED: हां हां हां हम्म
CAUSE: Model trained on longer segments. Short filler words/
       backchannels confuse it — it tries to generate more content.

CATEGORY 4: WORD DELETION / INSERTION (Moderate)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Model drops or adds words, changing sentence structure.

Examples:
  [2]  REF: ना चुनना      → PRED: चुनना  (dropped 'ना')
  [5]  REF: बोलते हैं कि  → PRED: बोलते कि (dropped 'हैं')
  [11] REF: घर के          → PRED: जरके   (merged+corrupted)
  [16] REF: बताने चाहेंगे → PRED: बताना जाएंगे
CAUSE: Fast conversational speech causes model to miss
       function words and grammatical markers.

CATEGORY 5: NAMED ENTITY / RARE WORD CONFUSION (Moderate)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Proper nouns, English loanwords, technical terms get corrupted.

Examples:
  [8]  REF: मिस्टेट       → PRED: देम स्टेट
  [8]  REF: सॉल्यूशन      → PRED: स्वालूसन
  [13] REF: स्वामी विवेकानंद → PRED: स्वामी वेकरनात
  [14] REF: एज्यूम        → PRED: कदिम
  [17] REF: रिस्टेट       → PRED: बाव री स्टेट
CAUSE: Rare proper nouns and English loanwords in Hindi
       are underrepresented in training data.
"""

print(taxonomy)

# ── Summary table ─────────────────────────────────────────────
print("\n=== CATEGORY FREQUENCY SUMMARY ===")
print(f"{'Category':<45} {'Count':>6} {'%':>6}")
print("-"*60)
categories = [
    ("Phonetic Confusion",              7,  28),
    ("Spelling Variation/Schwa",        6,  24),
    ("Short Utterance Hallucination",   5,  20),
    ("Word Deletion/Insertion",         4,  16),
    ("Named Entity/Rare Word",          3,  12),
]
for cat, count, pct in categories:
    print(f"{cat:<45} {count:>6} {pct:>5}%")
print("-"*60)
print(f"{'TOTAL':<45} {'25':>6} {'100':>5}%")
```

```python
# ── Part f: Proposed Fixes for Top 3 Error Types ─────────────

fixes = """
╔══════════════════════════════════════════════════════════════╗
║         TOP 3 ERROR TYPES — PROPOSED FIXES                  ║
╚══════════════════════════════════════════════════════════════╝

FIX 1: SHORT UTTERANCE HALLUCINATION
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Problem : Model hallucinates on utterances < 3 words
Fix     : Post-processing rule — if audio duration < 1.5 sec,
          use no_speech_threshold to suppress output OR
          limit generation length based on audio duration.
Specific: Set max_new_tokens = max(5, int(duration * 3))
          This prevents model from generating 10 words for 0.5s audio.
Not just more data — this is a decoding strategy fix.

FIX 2: PHONETIC CONFUSION
━━━━━━━━━━━━━━━━━━━━━━━━━
Problem : Similar consonants confused (स/श, त/ट, ब/व)
Fix     : Add data augmentation during training —
          pitch shifting, speed perturbation, additive noise.
          This forces model to learn robust phoneme boundaries
          rather than memorizing clean audio patterns.
Specific: Use audiomentations library with:
          - PitchShift(min=-2, max=2 semitones)
          - TimeStretch(min=0.9, max=1.1)
          - AddGaussianNoise(min=0.001, max=0.01)

FIX 3: SPELLING VARIATION / SCHWA DELETION
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Problem : सपने vs सप्ने, फिलहाल vs फिलाल — both valid
Fix     : Text normalization in post-processing.
          Normalize both reference and prediction before
          computing WER. Also standardize training labels
          using a Hindi text normalizer.
Specific: Use indic-nlp-library to normalize Devanagari
          before WER computation — this stops penalizing
          valid alternate spellings.
"""
print(fixes)
```

```python
# ── Part g: IMPLEMENT FIX 1 — Short Utterance Hallucination ──
import torch

# Find short utterance samples from test set
short_samples = []
for sample in test_dataset:
    ref = sample["text"].strip()
    word_count = len(ref.split())
    duration = len(sample["audio"]["array"]) / 16000
    if word_count <= 3 and duration < 2.0:
        short_samples.append({
            "audio_array": sample["audio"]["array"],
            "reference"  : ref,
            "duration"   : duration,
            "word_count" : word_count
        })
    if len(short_samples) >= 15:
        break

print(f"Found {len(short_samples)} short utterance samples\n")
print("=== BEFORE vs AFTER FIX ===\n")

before_preds = []
after_preds  = []
refs         = []

for s in short_samples:
    audio_array = s["audio_array"]
    ref         = s["reference"]
    duration    = s["duration"]

    inputs = finetuned_processor.feature_extractor(
        audio_array, sampling_rate=16000, return_tensors="pt"
    ).input_features.to("cuda")

    # ── BEFORE: no length limit ───────────────────────────────
    with torch.no_grad():
        pred_before_ids = finetuned_model.generate(
            inputs,
            language="hi",
            task="transcribe"
        )
    pred_before = finetuned_processor.tokenizer.batch_decode(
        pred_before_ids, skip_special_tokens=True
    )[0]

    # ── AFTER FIX: limit tokens based on duration ─────────────
    max_tokens = max(5, int(duration * 3))
    with torch.no_grad():
        pred_after_ids = finetuned_model.generate(
            inputs,
            language="hi",
            task="transcribe",
            max_new_tokens=max_tokens
        )
    pred_after = finetuned_processor.tokenizer.batch_decode(
        pred_after_ids, skip_special_tokens=True
    )[0]

    before_preds.append(pred_before)
    after_preds.append(pred_after)
    refs.append(ref)

    print(f"Duration: {duration:.1f}s | Words: {s['word_count']}")
    print(f"  REF   : {ref}")
    print(f"  BEFORE: {pred_before}")
    print(f"  AFTER : {pred_after}")
    print()

# ── WER comparison ────────────────────────────────────────────
wer_before = wer_metric.compute(predictions=before_preds, references=refs)
wer_after  = wer_metric.compute(predictions=after_preds,  references=refs)

print("="*55)
print("  SHORT UTTERANCE FIX — BEFORE vs AFTER WER")
print("="*55)
print(f"{'Condition':<35} {'WER':>10}")
print("-"*55)
print(f"{'Before fix (free generation)':<35} {wer_before*100:>9.2f}%")
print(f"{'After fix (length-limited)':<35} {wer_after*100:>9.2f}%")
print("-"*55)
improvement = (wer_before - wer_after) / wer_before * 100
print(f"Improvement on short utterances: {improvement:.1f}%")
print("="*55)
```
